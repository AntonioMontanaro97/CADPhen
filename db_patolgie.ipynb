```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
import statsmodels as stm
import missingno as msno
import random as rn
```

```python
#Leggo il dataset delle patologie
#La struttura sarà ID_PAZ PATOLOGIA_1 PATOLOGIA_2 PATOLOGIA_3...
dataSet = pd.read_csv("C:/Users/Antonio Montanaro/Desktop/Tesi Magistrale/File Memory/dataset_Patologie.csv")
```

```python
#Mantengo solo quelle righe per cui la colonna "RepeatInstrument" != NAN
dataSet = dataSet.loc[dataSet["Repeat Instrument"].notnull()]
del dataSet["Unnamed: 0"]
```

```python
# Costruisco il dataset in questo modo sapendo che al massimo sono state diagnosticate 3 patologie  (è possibile 
# trasformare tutto in funzione soprattutto se dovessero aumentare le patologie)
db_pato = pd.DataFrame(columns=["ID_PAZ", "PATO_0", "PATO_1", "PATO_2"])
id_paz = set(dataSet["ID PAZIENTE"])
for j in id_paz:
    db = dataSet.loc[dataSet["ID PAZIENTE"]==j]
    n_pato = db.shape[0]
    db_pato.at[j,"ID_PAZ"] = j
    if n_pato == 1:
        db_pato.at[j, "PATO_0"] = db["Patologia"].loc[db["Repeat Instance"] == 1.0]
    if n_pato == 2:
        db_pato.at[j,"PATO_0"] = db["Patologia"].loc[db["Repeat Instance"] == 1.0]
        db_pato.at[j,"PATO_1"] = db["Patologia"].loc[db["Repeat Instance"] == 2.0]
    if n_pato == 3:
        db_pato.at[j,"PATO_0"] = db["Patologia"].loc[db["Repeat Instance"] == 1.0]
        db_pato.at[j,"PATO_1"] = db["Patologia"].loc[db["Repeat Instance"] == 2.0]
        db_pato.at[j,"PATO_2"] = db["Patologia"].loc[db["Repeat Instance"] == 3.0]
```

```python
#Salvo il file
db_pato.to_csv("C:/Users/Antonio Montanaro/Desktop/Tesi Magistrale/File Memory/dataset_Patologie.csv")
```
